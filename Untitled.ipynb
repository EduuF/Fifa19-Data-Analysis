```python
# -*- coding: utf 8


from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('seaborn-colorblind')
plt.ion()
```

```python
df = pd.read_csv('fifa19.csv')
```

# Tratamento dos dados

```python
#Setando o ID como index
df = df.set_index('ID')

#Tirando coluna que não possui nome
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)

#Tirando colunas que possuem links 
df = df.drop("Flag",1)
df = df.drop("Club Logo",1)
df = df.drop("Real Face",1)
df = df.drop("Photo",1)
df = df.drop("Value",1)

#Tirando os caracteres '€', 'M' e 'K' da coluna 'Value' e 'Wage'
    
df['Wage'] = df['Wage'].str.replace('€', '')
df['Wage'] = df['Wage'].str.replace('K', '')
df['Wage'] = df['Wage'].str.replace('M', '')

#df['Value'] = df['Value'].str.replace('€', '')
#df['Value'] = df['Value'].str.replace('K', '')
#df['Value'] = df['Value'].str.replace('M', '')
    
#Fazendo o casting das colunas 'Value' e 'Wage' para float
#df['Value'] = df['Value'].astype(float)
df['Wage'] = df['Wage'].astype(float)
```

# Teste de hipótese/IC

### Qual o IC dos salários dos 100 jogadores com chutes mais potentes?

```python
plt.hist(df['ShotPower'], bins=10, edgecolor='k',density = True)
plt.xlabel("Força do chute")
plt.ylabel("% total de jogadores")
plt.axvline(df['ShotPower'].mean(), color='r', linewidth=3)
plt.title('Força do chute de todos os jogadores e média')
plt.show()
```

```python
df_shot_wage = df[['Wage' ,'ShotPower']]
df_shot_wage = df_shot_wage.nlargest(100,['ShotPower'])
plt.bar(df_shot_wage['ShotPower'], df_shot_wage['Wage'])
plt.title('Relação Salário vs Força do chute')
plt.xlabel("Força do chute")
plt.ylabel("Salário")
plt.show()
```

```python
def calcula_IC_salario(df, feature, top_x): #feature = feature a ser comparada c/ salario
                                            #top_x = seleciona os x jogadores com maior pontuação na feature
        
    df2 = df[['Wage' ,feature]] #Seleciona colunas de interesse
    df2 = df.nlargest(top_x,[feature]) #Os 100 jogadores com chutes mais potentes
    salario = df2['Wage']
    s = np.std(salario, ddof=1) ##1 grau de liberdade
    s_sobre_raiz_n = (s / np.sqrt(len(salario)))
    media = np.mean(salario)
    inferior = media-(1.96 * s_sobre_raiz_n)
    superior = media+(1.96 * s_sobre_raiz_n)
    return [ inferior, superior ]
```

```python
shotPower_Wage_IC = calcula_IC_salario(df, 'ShotPower', 100)
shotPower_Wage_IC
```
